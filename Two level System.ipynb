```python
import numpy as np
import matplotlib.pyplot as plt

def combination(n,x):
    return np.math.factorial(n)/(np.math.factorial(x)*np.math.factorial(n-x))

N=100
N1 = np.arange(N+1)
E = np.arange(-N,N+1,2)
omega = np.zeros(len(N1))

for i in range(len(N1)):
    omega[i] = combination(N,N1[i])
    
S = np.log(omega) #consider k_B = 1

plt.plot(E/N, S/N)
plt.xlabel("E/N$\epsilon$")
plt.ylabel("S/Nk$_B$")
plt.title("Plot of S/Nk$_B$ vs. E/N$\epsilon$ for N=100")
plt.grid(True)
plt.show()

def Tplot(N):
    N1 = np.arange(N+1)
    E = np.arange(-N,N+1,2)
    omega = np.zeros(len(N1))

    for i in range(len(N1)):
        omega[i] = combination(N,N1[i])
    S=np.log(omega)
    dE = abs(E[2]-E[1])
    T_inv = np.gradient(S,dE) #from thermodynamic relation, 1/T = dS/dE
    T=1/(T_inv)
    return plt.plot(T, E/N, 'r--')

fig = plt.figure(figsize=(20,8))

ax1=plt.subplot(1,3,1)
a = Tplot(10)
plt.xlabel("k$_B$T/$\epsilon$")
plt.ylabel("E/N$\epsilon$")
plt.title("Plot of E/N$\epsilon$ vs k$_B$T/$\epsilon$ for N=10")
plt.grid(True)

ax2=plt.subplot(1,3,2)
b = Tplot(50)
plt.xlabel("k$_B$T/$\epsilon$")
plt.ylabel("E/N$\epsilon$")
plt.title("Plot of E/N$\epsilon$ vs k$_B$T/$\epsilon$ for N=50")
plt.grid(True)

ax3=plt.subplot(1,3,3)
c = Tplot(100)
plt.xlabel("k$_B$T/$\epsilon$")
plt.ylabel("E/N$\epsilon$")
plt.title("Plot of E/N$\epsilon$ vs k$_B$T/$\epsilon$ for N=100")
plt.grid(True)
plt.show()
```
