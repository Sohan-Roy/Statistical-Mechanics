# Binomial Distribution

The probability of getting x successes in N trials is given by:
$$P(x) =  ^NC_x p^x q^{N-x}$$

Plot P(x) vs. x for N=20 and N=200, with p=q=$\frac{1}{2}$. Also plot ln P(x) vs. x and explain. 

```python
import numpy as np
import matplotlib.pyplot as plt

def combination(x,n):
    return np.math.factorial(n)/(np.math.factorial(x)*np.math.factorial(n-x))
def prob(x,n,p):
    return combination(x,n)*(p**x)*((1-p)**(n-x))

p=0.5
n1,n2=20,200
x1=np.arange(0,n1+1,1)
P1=np.zeros(len(x1))
for i in range(len(x1)):
    P1[i] = prob(x1[i],n1,p)
    
x2=np.arange(0,n2+1,1)
P2=np.zeros(len(x2))
for i in range(len(x2)):
    P2[i] = prob(x2[i],n2,p)
    

#Plotting
plt.subplot(2,2,1)
plt.plot(x1,P1,'r')
plt.xlabel("x")
plt.ylabel("P(x)")
plt.title(f"Plot of P(x) vs. x for N={n1}")

plt.subplot(2,2,2)
plt.plot(x1,np.log(P1),'g')
plt.xlabel("x")
plt.ylabel("ln P(x)")
plt.title(f"Plot of ln P(x) vs. x for N={n1}")

plt.subplot(2,2,3)
plt.plot(x2,P2,'r')
plt.xlabel("x")
plt.ylabel("P(x)")
plt.title(f"Plot of P(x) vs. x for N={n2}")

plt.subplot(2,2,4)
plt.plot(x2,np.log(P2),'g')
plt.xlabel("x")
plt.ylabel("ln P(x)")
plt.title(f"Plot of ln P(x) vs. x for N={n2}")

plt.tight_layout()
plt.show()
```
